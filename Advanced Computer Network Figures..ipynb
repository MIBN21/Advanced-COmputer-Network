```python
import matplotlib.pyplot as plt
import numpy as np

# Define the age groups and their respective data
age_groups = ["Less than 7", "7-14 years", "15-24 years", "25-34 years", "35-49 years", "50-59 years", "60 years and above", "All residents"]
data_2017 = [53, 96, 99, 99, 97, 85, 42, 84]
data_2018 = [77, 96, 100, 100, 98, 88, 55, 87]
data_2019 = [80, 99, 100, 100, 99, 95, 58, 89]

# Set the width of the bars and the gap
bar_width = 0.2
bar_gap = 0.1  # Adjust the gap size

# Set the positions of the bars on the x-axis
x = np.arange(len(age_groups))
x_2017 = x - bar_width - bar_gap
x_2018 = x
x_2019 = x + bar_width + bar_gap

# Create subplots to show data for each year side by side
fig, ax = plt.subplots(figsize=(12, 6))

bar_2017 = plt.bar(x_2017, data_2017, bar_width, label='2017', color='b', alpha=0.7)
bar_2018 = plt.bar(x_2018, data_2018, bar_width, label='2018', color='g', alpha=0.7)
bar_2019 = plt.bar(x_2019, data_2019, bar_width, label='2019', color='r', alpha=0.7)

# Increase the gap between "Less than 7" and "7-14 years" age groups
x_tick_positions = x

# Set x-axis labels and positions with horizontal rotation
plt.xlabel('Age Group', fontweight='bold')
plt.xticks(x_tick_positions, age_groups, rotation=0, fontweight='bold')

# Set y-axis label
plt.ylabel('Percentage of Internet Users', fontweight='bold')

# Set the title
plt.title('Proportion of Internet Users by Age Group, 2017-2019', fontweight='bold')

# Add a legend to the top left
plt.legend(loc='upper left')

# Display the percentages above the bars, centered within each age group
for i, val, label in zip(x_2017, data_2017, age_groups):
    plt.text(i + bar_width/2, val + 1, f'{val}%', ha='center', fontweight='bold')
for i, val in zip(x_2018, data_2018):
    plt.text(i + bar_width/2, val + 1, f'{val}%', ha='center', fontweight='bold')
for i, val in zip(x_2019, data_2019):
    plt.text(i + bar_width/2, val + 1, f'{val}%', ha='center', fontweight='bold')

# Adjust the y-axis limits
plt.ylim(0, 105)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['General Population', 'Seniors']
sizes = [49, 67]
colors = ['lightcoral', 'lightblue']
explode = (0.1, 0)  # Explode the first slice (Yes)

# Calculate values A and B
total_users = 1848
A = sizes[0] / 100 * total_users
B = sizes[1] / 100 * total_users

# Create the pie chart
fig, ax = plt.subplots(figsize=(6, 6))
patches, texts, autotexts = plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Make labels and autotexts bold
for text in texts:
    text.set_fontweight('bold')

for autotext in autotexts:
    autotext.set_fontweight('bold')

# Modify the text within the pie chart segments
autotexts[0].set_text(f'49% ({int(A)})')
autotexts[1].set_text(f'78.6% ({int(B)})')

# Add a legend without a title, and shift it to the top-left and vertically downwards
legend = ax.legend(labels, loc='upper left', bbox_to_anchor=(0.0, 0.95))
for text in legend.get_texts():
    text.set_fontweight('bold')

# Add the title to the pie chart
plt.title(' only', weight='bold')

# Show the pie chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

countries = [
    "Switzerland", "Ireland", "Netherlands", "Germany", "Australia", "Israel",
    "Denmark", "Sweden", "Belgium", "Singapore", "United States", "Hong Kong",
    "Taiwan", "Norway", "United Kingdom", "Finland", "New Zealand", "South Korea",
    "United Arab Emirates", "Czech Republic", "Spain", "Portugal", "Austria", "France",
    "Canada", "Japan", "Greece", "Slovakia", "Hungary", "Italy", "Saudi Arabia", "Poland"
]
rankings = list(range(1, 33))

# Set colors
colors = ["red" if country == "Singapore" else "blue" for country in countries]

# Create the figure
fig, ax = plt.subplots()

# Set the figure size
fig.set_size_inches(10, 8)

# Create horizontal bar chart
bars = ax.barh(countries, rankings, color=colors)

# Highlight Singapore by making its text red and bold
for i in range(len(countries)):
    if countries[i] == "Singapore":
        bars[i].set_color("red")
        ax.text(rankings[i] + 1, i, str(rankings[i]), va="center", fontsize=12, fontweight="bold", color="red")
    else:
        ax.text(rankings[i] + 1, i, str(rankings[i]), va="center", fontsize=12, fontweight="normal", color="blue")

# Add labels and title
ax.set_xlabel("Ranking", fontsize=12, fontweight="bold")
ax.set_ylabel("Country", fontsize=12, fontweight="bold")
ax.set_title("World Index of Healthcare Innovation Rankings", fontsize=14, fontweight="bold")

# Invert the y-axis to display the highest rank at the top
ax.invert_yaxis()

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# Define a font for the annotations
prop = fm.FontProperties(family="Arial", weight="bold")

# Age structure data
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
age_0_14 = [13.80, 13.53, 13.13, 12.71, 12.44, 12.29, 12.21, 12.21, 12.15, 11.99, 11.84]
age_15_64 = [78.33, 78.28, 78.37, 78.36, 77.99, 77.33, 76.53, 75.60, 74.69, 73.88, 73.04]
age_65_and_older = [7.88, 8.20, 8.51, 8.92, 9.56, 10.37, 11.26, 12.20, 13.15, 14.13, 15.12]

# Create the line chart
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(years, age_0_14, label='0-14 years', marker='o')
plt.plot(years, age_15_64, label='15-64 years', marker='o')
plt.plot(years, age_65_and_older, label='65 years and older', marker='o')

# Add labels and title
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Percentage of Population', fontweight='bold')
plt.title('Age Structure in Singapore (2012-2022)', fontweight='bold')

# Add a legend with bold font
legend = plt.legend()
for label in legend.get_lines():
    label.set_linestyle('-')
    label.set_linewidth(2)
legend.get_texts()[0].set_text('0-14 years')
legend.get_texts()[1].set_text('15-64 years')
legend.get_texts()[2].set_text('65 years and older')
for text in legend.get_texts():
    text.set_fontproperties(prop)

# Annotate data points for the year 2012 and 2022
plt.annotate(f'{age_0_14[0]:.2f}%', (years[0], age_0_14[0]), textcoords="offset points", xytext=(0, 10), ha='center', fontproperties=prop)
plt.annotate(f'{age_15_64[0]:.2f}%', (years[0], age_15_64[0]), textcoords="offset points", xytext=(0, 10), ha='center', fontproperties=prop)
plt.annotate(f'{age_65_and_older[0]:.2f}%', (years[0], age_65_and_older[0]), textcoords="offset points", xytext=(0, 10), ha='center', fontproperties=prop)

plt.annotate(f'{age_0_14[-1]:.2f}%', (years[-1], age_0_14[-1]), textcoords="offset points", xytext=(0, 10), ha='center', fontproperties=prop)
plt.annotate(f'{age_15_64[-1]:.2f}%', (years[-1], age_15_64[-1]), textcoords="offset points", xytext=(0, 10), ha='center', fontproperties=prop)
plt.annotate(f'{age_65_and_older[-1]:.2f}%', (years[-1], age_65_and_older[-1]), textcoords="offset points", xytext=(0, 10), ha='center', fontproperties=prop)

# Show the plot
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the age groups and their respective data
age_groups = ["0 - 14", "15 - 64", "65 & Above"]
data_acute_hospitals = [104.1, 81.4, 302.7]
data_psychiatric_hospitals = [0.2, 2.3, 1.6]
data_community_hospitals = [0.0, 1.2, 22.7]

# Set the width of the bars and the gap
bar_width = 0.2
bar_gap = 0.1  # Adjust the gap size

# Set the positions of the bars on the x-axis
x = np.arange(len(age_groups))
x_acute_hospitals = x - bar_width - bar_gap
x_psychiatric_hospitals = x
x_community_hospitals = x + bar_width + bar_gap

# Create subplots to show data for each hospital type side by side
fig, ax = plt.subplots(figsize=(12, 6))

bar_acute_hospitals = plt.bar(x_acute_hospitals, data_acute_hospitals, bar_width, label='Acute Hospitals', color='b', alpha=0.7)
bar_psychiatric_hospitals = plt.bar(x_psychiatric_hospitals, data_psychiatric_hospitals, bar_width, label='Psychiatric Hospitals', color='g', alpha=0.7)
bar_community_hospitals = plt.bar(x_community_hospitals, data_community_hospitals, bar_width, label='Community Hospitals', color='r', alpha=0.7)

# Set x-axis labels and positions with horizontal rotation
plt.xlabel('Age Group', fontweight='bold')
plt.xticks(x, age_groups, rotation=0, fontweight='bold')

# Set y-axis label
plt.ylabel('Hospital Admissions', fontweight='bold')

# Set the title
plt.title('Hospital Admissions by Age Group and Hospital Type (Rate per 1000 Population)', fontweight='bold')

# Add a legend to the top left
plt.legend(loc='upper left')

# Display the values directly above the bars, centered within each age group
for i, val in zip(x_acute_hospitals, data_acute_hospitals):
    plt.text(i + bar_width / 2, val + 10, f'{val}', ha='center', va='bottom', fontweight='bold')
for i, val in zip(x_psychiatric_hospitals, data_psychiatric_hospitals):
    plt.text(i + bar_width / 2, val + 0.5, f'{val}', ha='center', va='bottom', fontweight='bold')
for i, val in zip(x_community_hospitals, data_community_hospitals):
    plt.text(i + bar_width / 2, val + 2, f'{val}', ha='center', va='bottom', fontweight='bold')

# Adjust the y-axis limits
plt.ylim(0, 350)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Time series data
years = list(range(2018, 2029))
revenue = [2.92, 3.04, 3.25, 4.38, 5.47, 6.56, 7.65, 8.70, 9.73, 10.73, 11.63]

# Create a line chart
plt.plot(years, revenue, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('Year', weight='bold')
plt.ylabel('Revenue (Billion U.S. Dollars)', weight='bold')
plt.title('Revenue of the Internet of Things (IoT) market in Singapore (2018-2028)', weight='bold')

# Add text labels for each data point
for i, txt in enumerate(revenue):
    plt.annotate(f'{txt:.2f}', (years[i], revenue[i]), textcoords="offset points", xytext=(0, 5), ha='center', weight='bold')

# Show the plot
plt.grid(True)
plt.show()
```
